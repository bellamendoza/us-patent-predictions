```python
import pandas as pd
import plotly_express as px
import plotly.io as pio
```

```python
data = pd.read_csv('../data/input.csv')
```

```python
data
```

### Distribution of utility patents based on CPC

```python
pio.renderers.default = "browser" 

# Get value counts of 'cpc_section' (raw counts)
cpc_counts = data['cpc_section'].value_counts()

cpc_mapping = {
    'A': 'Human Necessities',
    'B': 'Operations & Transport',
    'C': 'Chemistry & Metallurgy',
    'D': 'Textiles & Paper',
    'E': 'Fixed Constructions',
    'F': 'Mech. Eng & Lighting',
    'G': 'Physics',
    'H': 'Electricity',
    'Y': 'New Tech Developments'
}

cpc_counts = cpc_counts.rename(cpc_mapping)

cpc_df = pd.DataFrame({'CPC Section': cpc_counts.index, 'Count': cpc_counts.values})

cpc_df = cpc_df.sort_values(by='Count', ascending=False)


fig = px.bar(cpc_df, x='CPC Section', y='Count')
fig.update_layout(
    xaxis=dict(
        tickfont=dict(size=20)  # Change 20 to any size you prefer
    )
)
fig.show()
```

### Distribution of patents based on federal agencies

```python
agency_counts = data['fedagency_name'].value_counts()
agency_df = pd.DataFrame({'Agency': agency_counts.index, 'Count': agency_counts.values})
agency_df = agency_df.sort_values(by='Count', ascending=False)

fig = px.bar(agency_df[:20], x='Agency', y='Count')
fig.update_layout(
    xaxis=dict(
        tickfont=dict(size=20)  # Change 20 to any size you prefer
    )
)
fig.show()
```
